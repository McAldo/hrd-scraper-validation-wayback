```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('hrd.db')

# Get list of all tables
tables = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;",
    conn
)['name'].tolist()

for table in tables:
    print(f"\n=== First 50 rows of `{table}` ===")
    df = pd.read_sql_query(f"SELECT * FROM {table} LIMIT 50;", conn)
    print(df)

conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('hrd.db')

# Get list of all tables
tables = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;",
    conn
)['name'].tolist()

for table in tables:
    print(f"\n=== Table `{table}` ===")

    # 1) List all columns via PRAGMA
    cursor = conn.execute(f"PRAGMA table_info({table});")
    cols = [row[1] for row in cursor.fetchall()]  # row[1] is the column name
    print("Columns:", cols)

    # 2) Show first 50 rows (if any)
    df = pd.read_sql_query(f"SELECT * FROM {table} LIMIT 50;", conn)
    if not df.empty:
        print("\nFirst 50 rows:")
        print(df)
    else:
        print("\n(no rows in table)")

conn.close()
```

## Importing from a SQLITE3 db to Pandas DF (no SQL alchemy required)

## Importing from a Database to Pandas DF through SQL Alchemy 

SQL Alchemy is a general interface with abstract SQL calls to a variety of Databases, including SQLITE3

```python
from sqlalchemy import create_engine
import pandas as pd

pd.set_option('display.max_columns', None)

engine = create_engine('sqlite:///hrd.db')
df_profiles = pd.read_sql_table('profiles', con=engine)

df_profiles


```

```python
## Close the connection with Alchemy
engine.dispose()
```
